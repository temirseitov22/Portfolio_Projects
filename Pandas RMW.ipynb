```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = 'Files/Sales/'
```

```python
for f in os.listdir(path):
    print(f)
```

```python
#Create empty df
df = pd.DataFrame() 

#Concatenate df with each csv file
for f in os.listdir(path):
    tmp = pd.read_csv(path + f, encoding='cp1252')
    df = pd.concat([df,tmp])
    
#show df
df
```

```python
df.describe()
```

```python
df.shape[0]
```

```python
#Data has 3505 rows but last 5 index numbers show not appropriate data
df.tail(5)
```

```python
#Reset index
df.reset_index(inplace=True)
df
```

```python
#Delete not appropriate datas
df.drop('index', axis = 1, inplace = True)
df
```

```python
#Show unique values of each column
df.nunique()
```

```python
#Delete data that we dont need in analysis
df.drop(['Department family','Entity', 'Description'], axis=1, inplace=True)
df.head(5)
```

```python
#Rename columns
df = df.rename(columns = {
                     'Expense type' : 'Expense_Type',
                    'Expense area' : 'Expense_Area',
                    'Transaction number' : 'Transaction_Numb',
                    'Supplier Postcode' : 'Postcode'
})
df.head()
```

```python
#Floats data formatting
pd.options.display.float_format = '${:,.0f}'.format
df.head()
```

```python
#Check columns types
df.dtypes
```

```python
#Converting Date type from 'object' to 'datetime64'
df['Date'] = pd.to_datetime(df.Date, format='%m%d%Y')
df.dtypes
```

```python
#Check for NaN data
df.isna().sum()
```

```python
#Delete NaN data
df.dropna(inplace=True)

df.isna().sum()
```

```python
#Create column with month names
df['Month'] = df.Date.dt.month_name()

#Create column with years
df['Year'] = df.Date.dt.year

df.head()
```

```python
total_per_year = df.groupby('Year', as_index=False).agg({'Amount' : 'sum'}).sort_values('Amount')
total_per_year
```

```python
sns.barplot(data=total_per_year,x='Year', y= 'Amount')

plt.ticklabel_format(style='plain', axis='y')
plt.title('Total Amount per Year\n', fontsize=14, weight='bold')
plt.show()
```

```python
sns.set(
    font_scale = 1,
    style = 'whitegrid',
    rc = {'figure.figsize' : (10,5)})
```

```python
pivot_year_amount = pd.pivot_table(df, values = 'Amount', index=['Year'], columns = 'Month', fill_value=0)
pivot_year_amount
```

```python
best_month_year = pd.pivot_table(df, values = 'Amount', index=['Month'], columns = 'Year', fill_value=0).idxmax()
best_month_year
```

```python
best_month_ever = df.groupby('Month')[['Amount']].sum().reset_index()
best_month_ever
```

```python
sns.barplot(data=best_month_ever,x='Month', y= 'Amount')

plt.ticklabel_format(style='plain', axis='y')
plt.title('Total Month Sales 2014-2021\n', fontsize=14, weight='bold')
plt.show()
```

```python
#Show best supplier
best_supplier = df.groupby('Supplier')[['Amount']].sum().sort_values('Amount', ascending=False).head(5).reset_index()
best_supplier
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(best_supplier.Amount, labels=best_supplier.Supplier, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```
