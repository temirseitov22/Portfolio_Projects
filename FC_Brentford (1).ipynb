```python
#Looking for Goalkeepers more than 0 rating and 'Low' wage level
#Sort df by Rating then Save percentage
squad_gk = epl_gk.query('Rating != "☆" and Wage_Level == "Low"')\
                 .sort_values(by = ["Rating","Save_perc"], ascending= False)

squad_gk
```

# Defending

```python
epl_def
```

```python
# Add wage column to defend players
epl_def = epl_def.merge(epl_wage)
```

```python
# Check that we have only DF position
epl_def.Pos.unique()
```

```python
# Looking for average defender stats with Low level wage
epl_def.query('Wage_Level == "Low"') \
        .mean()
```

```python
# Create filter for defender criteria
# using above given stats
mid_median = 'Pressure >= 173 & Blocks >=31 & Yellow_C <= 3 & Min >= 1431 & Wage_Level == "Low" '
```

```python
#Look for defenders with the highest pressure then success blocks percentage
fullback = epl_def.query(mid_median)\
                  .sort_values(["Pressure","Succ_perc"], ascending=False)\
                    .head(5)
fullback
```

```python
#Look DF who best in winning the ball
winner = epl_def.query(mid_median)\
                  .sort_values(["Succ_perc","Blocks"], ascending=False)\
                  .head(5)

winner
```

```python
#Look DF who help attacking
wingback = epl_def.query(mid_median)\
                  .sort_values("Goal_Assist", ascending=False)\
                  .head(5)

wingback
```

```python
# Concatenate selected defend players
defender = pd.concat([winner, fullback, wingback], ignore_index=True, sort=False)
```

```python
# Select 8 defenders with the highest rating and blocks
squad_def = defender.drop_duplicates().sort_values(["Rating","Blocks"], ascending=False).head(8)
squad_def
```

# Midfielder

```python
epl_mf
```

```python
# Add wage columns to MF table
epl_mf = epl_mf.merge(epl_wage)
```

```python
# Check that we have only MF related positions
epl_mf.Pos.unique()
```

```python
# Looking for average midfielder stats with Low level wage
epl_mf.query('Wage_Level == "Low"') \
      .mean()
```

```python
# Create filter for defender criteria using above given stats
Median_mf = 'Min >= 1343 & Passes >= 487 & Goal_Assist >= 2 & Key_pass >=14 & Wage_Level =="Low"'
```

```python
#Looking for general MF
gen_mf = epl_mf.query(Median_mf)\
               .sort_values(["Passes","Pass_perc"], ascending=False)\
               .head(5)

gen_mf
```

```python
# Looking for attacking MF
attack_mf = epl_mf.query(Median_mf)\
                  .sort_values(["Goal_Assist", "Key_pass"], ascending=False)\
                  .head(5)

attack_mf
```

```python
# Concatenate selected MF players
all_mf = pd.concat([gen_mf, attack_mf], ignore_index=True, sort=False)
```

```python
# Select 5 MF players with highest Rating and Goal+Pass
squad_mf = all_mf.drop_duplicates()\
                 .sort_values(["Goal_Assist", "Rating"], ascending=False,ignore_index=True)\
                 .head(5)
squad_mf
```

# Attacking

```python
epl_attack
```

```python
epl_attack.dtypes
```

```python
# Add wage columns to attack table
epl_attack = epl_attack.merge(epl_wage)
```

```python
# Check we have only FW related positions
epl_attack.Pos.unique()
```

```python
# Look for average forward stats with Low level wage
epl_attack.query('Wage_Level == "Low"') \
          .mean()
```

```python
# Create filter for attacking players criteria using above given stats
attack_median = 'Min >= 1293 and Shots >= 30 and Shot_create >= 34 and Goal_create >= 4 and Age < 30 and Wage_Level == "Low"'
```

```python
#Looking for strikers
striker = epl_attack.query(attack_median)\
                    .sort_values(["Goals", "Goal_create"], ascending=False)\
                    .head(5)
striker
```

```python
# Looking for attack creators
creator = epl_attack.query(attack_median)\
                    .sort_values(["Goal_create", "Shot_create"], ascending=False)\
                    .head(5)
creator
```

```python
# Looking for assistance forwards
assist_attacker = epl_attack.query(attack_median)\
                            .sort_values(["Assists", "Goal_create"], ascending=False)\
                            .head(5)
assist_attacker
```

```python
# Concatenate selected FW players
forwards = pd.concat([striker, creator, assist_attacker], ignore_index=True, sort=False)
```

```python
# Select 6 FW players with highest Goaling stats
squad_attack = forwards.drop_duplicates()\
                       .sort_values(["Goal_create","Goals"], ascending=False)\
                       .head(6)
squad_attack
```

```python
#Create df with all selected players

#Collect all positions selected players to one df
squad_all = pd.concat([squad_gk, squad_def, squad_mf, squad_attack], ignore_index=True, sort=False)

#Give columns names
squad_all = squad_all[['Player','Pos','Nation','Team','Age','Min','Year_Wage','Rating']]

#Assign GK position instead of null value
squad_all = squad_all.fillna('GK')

#Show whole squad
squad_all
```
